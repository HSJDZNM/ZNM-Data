```python
from bs4 import BeautifulSoup
import requests
import csv

url = "https://www.basketball-reference.com/leagues/NBA_2022_games-october.html"
r = requests.get(url)
html = r.text
print(html)
```

```python
# 创建BeautifulSoup对象
soup = BeautifulSoup(html, 'html.parser')

tr_list = []

# 查找table标签
tbody = soup.find('tbody')

# 获取到每个tr
for tr in tbody:
    if tr != "\n":
        tr_list.append(tr)
    
del tr_list[0]
del tr_list[1]
print(tr_list[1])
```

```python
all_data_list = []

# 遍历所有tr标签
for tr in tr_list:
    data_list = []

    print(tr)
    soup = BeautifulSoup(str(tr), 'html.parser')

    # 找到所有th标签的值
    th = tr.find('th')
    th_value = th.get_text()
    print("th标签的值:", th_value)
    data_list.append(th_value)
    
    # 找到所有td标签的值
    td_list = tr.find_all('td')
    for td in td_list:
        td_value = td.get_text()
        print("td标签的值:", td_value)
        data_list.append(td_value)
    
    all_data_list.append(data_list)

print(all_data_list)
```

```python
csv_file = 'NBA_data.csv'

# 打开CSV文件并写入数据
with open(csv_file, 'w', newline='', encoding='utf8') as f:
    writer = csv.writer(f)
    for data in all_data_list:
        writer.writerow(data)
print(f'Data has been written to {csv_file}')
```
