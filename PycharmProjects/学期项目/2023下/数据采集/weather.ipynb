```python
import re
from bs4 import BeautifulSoup
import csv
```

```python
file_path = "C:\\Users\\dell\\Desktop\\weather.html"

try:
    with open(file_path, 'r', encoding='gb2312',) as file:
        html_content = file.read()
        print(html_content)
except FileNotFoundError:
    print(f"文件 '{file_path}' 未找到")
except Exception as e:
    print(f"发生错误: {str(e)}")
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
tbody = soup.find('tbody')
print(tbody)
```

```python
td = soup.find_all("td")
print(td)
```

```python
# 将原始列表分为每四个元素一组的子列表
chunked_lists = [td[i:i+4] for i in range(0, len(td), 4)]

# 指定CSV文件名
csv_file = 'data.csv'
datalst = []

# 输出结果
for chunk in chunked_lists:
    print(chunk)
    data = [element.get_text() for element in chunk]
    print(data)
    datalst.append(data)

# 打开CSV文件并写入数据
with open(csv_file, 'w', newline='', encoding='utf8') as f:
    writer = csv.writer(f)
    for data in datalst:
        writer.writerow(data)
print(f'Data has been written to {csv_file}')

```
