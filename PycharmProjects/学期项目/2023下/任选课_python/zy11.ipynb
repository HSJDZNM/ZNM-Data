```python
import turtle

turtle.begin_fill()  # 开始填充
turtle.fillcolor("red")  # 设置填充颜色为红色

for _ in range(8):
    turtle.forward(100)
    turtle.right(45)

turtle.end_fill()  # 结束填充

turtle.done()
```

```python
import turtle

def draw_star(size, points):
    angle = 360 / points  # 计算每个尖角之间的角度

    for _ in range(points):
        turtle.forward(size)
        turtle.right(180 - angle)

# 设置画笔速度和起始位置
turtle.speed(10)
turtle.penup()
turtle.goto(-50, 0)
turtle.pendown()

# 画一个大小为100，有8个尖角的星星
draw_star(100, 16)

turtle.done()
```

```python
import tkinter as tk
from PIL import Image, ImageTk

class GIFViewer:
    def __init__(self, master, gif_path):
        self.master = master
        self.master.title("显示GIF图片")

        # 打开GIF文件
        self.gif = Image.open(gif_path)

        # 创建PhotoImage对象
        self.gif_photo = ImageTk.PhotoImage(self.gif)

        # 在Label中显示GIF图片
        self.label = tk.Label(self.master, image=self.gif_photo)
        self.label.photo = self.gif_photo  # 存储为窗口对象的属性，以避免被垃圾回收
        self.label.pack()

if __name__ == "__main__":
    root = tk.Tk()

    # 替换为你的GIF图片路径
    gif_path = "C:\\Users\\dell\\Pictures\\Camera Roll\\R.gif"
    
    app = GIFViewer(root, gif_path)
    
    # 移动主循环到类的外部
    root.mainloop()
```

```python
import tkinter as tk
from PIL import Image, ImageTk

class MovingImage:
    def __init__(self, master, gif_path):
        self.master = master
        self.master.title("移动图片")

        # 打开GIF文件
        self.gif = Image.open(gif_path)
        self.gif_width, self.gif_height = self.gif.size

        # 创建Canvas组件
        self.canvas = tk.Canvas(self.master, width=self.gif_width, height=self.gif_height)
        self.canvas.pack()

        # 创建PhotoImage对象
        self.gif_photo = ImageTk.PhotoImage(self.gif)

        # 在Canvas上显示GIF图片
        self.image_item = self.canvas.create_image(self.gif_width // 2, self.gif_height // 2, image=self.gif_photo)

        # 记录原始图片位置
        self.original_position = (self.gif_width // 2, self.gif_height // 2)

        # 设置移动参数
        self.move_distance = 5
        self.direction = 1  # 1表示向下移动，-1表示向上移动
        self.move_count = 0  # 记录移动次数

        # 启动移动循环
        self.move_image()

    def move_image(self):
        # 获取当前图片的坐标
        x, y = self.canvas.coords(self.image_item)

        # 检查是否超出上下边界，如果是，则改变移动方向
        if y <= 0 or y >= self.gif_height:
            self.direction *= -1
            self.move_count += 1

            # 当移动四次后回到原位并停止运动
            if self.move_count >= 4:
                self.canvas.coords(self.image_item, *self.original_position)  # 将图片移回原始位置
                return

        # 移动图片
        self.canvas.move(self.image_item, 0, self.move_distance * self.direction)

        # 设置定时器，继续移动
        self.master.after(2, self.move_image)

if __name__ == "__main__":
    root = tk.Tk()

    # 替换为你的GIF图片路径
    gif_path = "C:\\Users\\dell\\Pictures\\Camera Roll\\R.gif"
    
    app = MovingImage(root, gif_path)

    root.mainloop()
```
